```r
#takes 55 seconds for my laptop
prices<-c()
volumes<-c()
nominals<-c()
for (i in 1:250) {
    ticker <- sampletickers[i]
    stock_data <- getSymbols(
        ticker,
        from = as.Date("2018-01-01"),
        to = as.Date("2021-06-01"),
        src = 'yahoo',
        warnings = FALSE,
        auto.assign = FALSE
    )
    stock_prices = stock_data[, 6]
    stock_volumes = stock_data[, 5]
    stock_nominals = stock_data[, 5] * stock_prices
    colnames(stock_prices) <- ticker
    colnames(stock_volumes) <- ticker
    colnames(stock_nominals) <- ticker
    prices <- cbind(prices, stock_prices)
    volumes<- cbind(volumes, stock_volumes)
    nominals <- cbind(nominals, stock_nominals)
}
```
